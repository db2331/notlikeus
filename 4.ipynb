```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('titanic.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
#plotting bar using x and y axis
plt.bar(data['Age'],data['Fare'])
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Fare per age")
plt.show()
```

```python
#plotting bar using x and y axis
plt.bar(data['Survived'],data['Age'])
plt.xlabel("Survived")
plt.ylabel("Age") 
plt.title("Survived per age")
plt.show()
```

```python
#create count plot to compare the distribution of age between survivors and non-survivors
sns.catplot(data=data,x='Survived',hue='Age',kind='count')
plt.title('Survival Distribution by Age')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```

```python
data.groupby("Sex")["Age"].mean()
```

```python
#box plot
sns.catplot(x="Sex",y="Age",data=data,kind="box")
plt.show()
```

```python
#bar plot
sns.catplot(x="Parch",y ="Survived",data=data,kind="bar")
plt.show()
```

```python
#violin plot
sns.catplot(data=data,x="Sex",y="Age",kind="violin")
plt.title('Age Distribution by Class(Violin Plot)')
plt.show()
```

```python
#strip plot
sns.catplot(data=data,x='Pclass',y='Age',kind='strip')
plt.title('Individual Ages by Class (Strip Plot)')
plt.show()
```

```python
#count of passengers by class
sns.catplot(data=data,x='Pclass',kind='count')
plt.title('Count of Passengers by Class')
plt.show()
```

```python
#count of passengers by sex
sns.catplot(data=data,x='Sex',kind='count')
plt.title('Count of Passengers by Sex')
plt.show()
```

```python
#box plot
sns.boxplot(x="Pclass",y="Age",data=data)
plt.show()
```

```python
#pair plot
sns.pairplot(data)
```

```python
sns.distplot(x=data['Pclass'])
plt.show()
```

```python
#violin plot
sns.violinplot(x=data['Sex'],y=data['Age'])
```

```python
#strip plot
sns.stripplot(x=data['Pclass'],y= data['Age'])
```

```python
# filter passengers who paid more than $500
high_fare_passengers = data[data['Fare'] > 5]
sns.countplot(data=high_fare_passengers, x='Survived')
plt.title('Survival Status of Passengers Who Paid More Than $5')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count of Passengers')
plt.show()
survival_rate = high_fare_passengers['Survived'].value_counts(normalize=True)
print(f"Survival Rate for Passengers Who Paid More Than $5:\n{survival_rate}")
```

**Analysis - The plot shows survival distribution, with red for non-survivors and green for survivors. If survival is high, it suggests wealthier passengers had better access to lifeboats. If balanced or low, other factors like demographics or ship location influenced survival. Here, the survival of passengers who paid more than 5 dollars is low.**

```python
#filter by gender the ones who survived and paid fare

paid_fare=data[data['Fare']>0]
sns.countplot(data=paid_fare,x='Sex',hue='Survived')
plt.title("filter by gender the ones who survived and paid fare")
plt.xlabel('Gender')
plt.ylabel('Fare')
plt.show()


# data_paid_fare = data[data['Fare'] > 0]
# sns.countplot(data=data_paid_fare,x="Sex",hue="Survived")
# plt.title("Survival Based on Gender Who Paid Fare")
# plt.xlabel("Gender")
# plt.ylabel("Count")
# plt.legend(title="Survived",labels=["No","Yes"])
# plt.show()
```

**Analysis - The plot shows survival distribution based on gender for passengers who paid a fare. The bars indicate that females had a higher survival rate compared to males. The lower survival rate of males suggests they had less access to lifeboats.**

```python
survived_paid_fare = data[(data['Survived']==1) & (data['Fare']>10)]
sns.countplot(data=survived_paid_fare,x="Sex")
plt.title("Survivors by Gender (Fare > 10)")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

**Analysis - The plot shows the gender distribution of passengers who survived and paid more than 10 dollars in fare. The number of female survivors is significantly higher than male survivors.The lower count of male survivors indicates that they faced greater challenges in securing lifeboat access.**

```python
#filter how many survived in embarked the stations
embarked_analysis=data[data['Survived']==1].groupby('Embarked').size()
sns.barplot(x=embarked_analysis.index,y=embarked_analysis.values,palette='viridis')
plt.title("Count of Survivors by Embarkation Point",fontsize=16)
plt.xlabel("Embarkation Point",fontsize=12)
plt.ylabel("Count of Survivors",fontsize=12)
plt.show()
```

**Analysis - The plot shows the count of survivors based on their embarkation point. Passengers who embarked from point 'S' had the highest number of survivors, possibly due to a higher number of wealthy passengers. Those from 'C' also had a notable count, while 'Q' had the fewest survivors. This suggests that embarkation location influenced survival, likely due to differences in class distribution and access to lifeboats.**

```python
data.at[336,'sex'] = 'Female'
```

```python
#scatter plot for length and depth
sns.scatterplot(x="bill_length_mm", y="bill_depth_mm",data=data,hue="species")
plt.title("Bill Length vs Bill Depth",size=20,color="red")
```

**Analysis - The scatter plot visualizes the relationship between bill length and bill depth across different species. The points are color-coded by species, showing how they cluster based on these features. Distinct groupings suggest that bill dimensions vary significantly among species. This separation indicates that bill length and depth are key characteristics in distinguishing species.**  

```python
#histogram plot for flipper length on x
sns.histplot(x ="flipper_length_mm", data=data,binwidth=3,kde=True)
plt.title("Flipper Length", size=20, color="red")
```

```python
#histogram plot for flipper length on y
sns.histplot(y = "flipper_length_mm", data = data)
plt.title("Flipper Length", size=20, color="red")
```

```python
#histogram of categories
sns.histplot(x = "flipper_length_mm", data = data, hue = "species")
plt.title("Flipper Length", size=20, color="red")
```

```python
#flipper length for 3 species
sns.barplot(x ="species", y="flipper_length_mm", data=data)
plt.title("Flipper Length for 3 Penguin Species", size=20, color="red")
```

```python
#flipper length for 3 species by length
sns.barplot(x = "species", y = "flipper_length_mm", data =data, hue = "sex")
plt.title("Flipper Length for 3 Penguin Species by Sex", size=20, color="red")
```

**Analysis - The bar plot shows the flipper length distribution for three penguin species, categorized by sex. The variation in flipper length across species is evident, with some species having noticeably longer flippers. Differences between male and female penguins are also highlighted.This suggests that both species and sex influence flipper length.**  

```python
#boxplot for species flipper length
sns.boxplot(x = "species", y = "flipper_length_mm", data =data)
plt.title("Flipper Length for 3 Penguin Species", size=20, color="red")
```

```python
#flipper length for 3 species with sex
sns.boxplot(x = "species", y = "flipper_length_mm", data =data, hue = "sex")
plt.title("Flipper Length for 3 Penguin Species by Sex", size=20, color="red")
```

```python
#violin plot for flipper length by species
sns.violinplot(x = "species", y = "flipper_length_mm", data = data)
plt.title("Flipper Length for 3 Penguin Species", size=20, color="red")
```

```python
#facet plot
sns.FacetGrid(data, col="island", row="sex").map(sns.histplot, "flipper_length_mm")
```

```python
#dist plot for above
sns.FacetGrid(data, col = "island", row = "sex").map(sns.distplot, "flipper_length_mm")
```

```python
#heatmap
num_penguins=data.select_dtypes(include=['float','int64'])
sns.heatmap(num_penguins.corr(),annot=True,cmap='coolwarm')
```

**Analysis - The heatmap displays the correlation between numerical features in the dataset. Darker shades indicate stronger correlations, with positive correlations in red and negative correlations in blue. Features like flipper length, bill length, and body mass may show strong relationships, suggesting patterns within species.**
