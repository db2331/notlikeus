Plot a box plot for distribution of age with respect to each gender along with the information about whether they survived
or not. (Column names: 'sex' and 'age')

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('titanic.csv')
data.head()
```

```python
#plotting a boxplot for age wrt sex and survived
sns.boxplot(x='Sex',y='Age',hue='Survived',data=data)
plt.title("Age Distribution by Gender and Survival Status(1-survived 0-not survived)")
plt.xlabel("Gender")
plt.ylabel("Age")
plt.show()
```

**Analysis - The survival rate of female is more in average age form 20 to 40. The unsurvival rate is more in males in age group between 20 to 40**

**Analysis - The boxplot visualizes the age distribution by gender and survival status. It shows that female survivors tend to be younger, with a lower median age compared to non-survivors. Male survivors also have a similar age distribution, though the median age for male non-survivors is slightly higher. This suggests that age played a role in survival, with younger passengers more likely to survive, and gender influencing access to lifeboats.**

```python
sns.barplot(x='Sex',y='Age',hue='Survived',data=data)
plt.title("Mean Age Distribution by Gender and Survival Status(1-survived 0-not survived)")
plt.xlabel("Gender")
plt.ylabel("Mean Age")
plt.show()
```

**Analysis - The bar plot shows the mean age distribution by gender and survival status. Female survivors have a lower mean age compared to female non-survivors, while male survivors tend to have a slightly higher mean age than male non-survivors. This suggests that younger passengers, particularly females, had a higher chance of survival.**
