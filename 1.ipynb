```python
#NORMALIZATION - MINMAXSCALER
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
df_norm=scalar.fit_transform(data[['Weight']])
df_norm_df = pd.DataFrame(df_norm, columns=['Weight_Scaled'])
data = pd.concat([data, df_norm_df], axis=1)
print(data)
```

```python
#NORMALIZATION - Z-SCORE
from sklearn.preprocessing import StandardScaler
sscalar=StandardScaler()
df_norm=sscalar.fit_transform(data[['Salary']])
df_norm_df = pd.DataFrame(df_norm, columns=['Sal_Scaled'])
data = pd.concat([data, df_norm_df], axis=1)
print(data)
```
